```python
### load modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# datareader
from pandas_datareader import wb
```

```python
# 
sns.set(style = "whitegrid", rc = {"figure.figsize": (10, 8)})
```

```python
### example data
ue = pd.read_csv('tolpp.csv')
ue
```

```python
sns.set(color_codes=True)
sns.lineplot(data = ue, x = "age", y = "tolerance", 
             hue = "id", 
             palette = 'crest')
```

```python
# Convert the year to integer for better labels
ue = ue.assign(year = ue.age.astype(int))

# Make a Greece-only dataset
greece = ue.query('id == 1653')

# Plot everything as grey, first
sns.lineplot(data = ue, x = "age", y = "tolerance",
             hue = "id", palette = ["grey"] * 16,
             legend = False)

# Then, plot Greece on top
sns.lineplot(data = greece, x = "age", y = "tolerance",
             hue = "id", palette = ["blue"], linewidth = 3)
```
